```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv('googleplaystore.csv')
df.isnull().sum()
```

```python
imputer = SimpleImputer(missing_values=np.nan, strategy = "mean")
imputer.fit(df.iloc[:,2:3])
df.iloc[:,2:3] = imputer.transform(df.iloc[:,2:3])
df.isnull().sum()
```

```python
df_nullChecked = df.dropna()
df_nullChecked.isnull().sum()
```

Q1) Name all the apps that are in Art and Design and rating above or 4 in ascending order

```python
df_pr = df_nullChecked[df_nullChecked['Category'] == 'ART_AND_DESIGN']
df_pr2 = df_pr[df_pr['Rating'] >= 4]
df_pr2.sort_values(by='Rating', ascending= True)
```

Q2)Which category has maximum rating

```python
df_nullChecked.groupby('Category').mean()['Rating'].sort_values(ascending=False)
```

```python
Category = {}
for name in df_nullChecked['Category'].unique():
    count = 0
    for i in df_nullChecked['Category']:
        if (i == name):
            count += 1
    Category[name]=count
Category
```

```python
df_nullChecked.groupby('Category').count()['Type'].sort_values(ascending = False)
```
